```python
likes = []
reposts = []
for post in posts:
    likes.append(post['likes']['count'])
    reposts.append(post['reposts']['count'])
```

```python
x_axes = []
for each in enumerate(likes):
    x_axes.append(each[0])
```

```python
import pandas as pd
df = pd.DataFrame({'Likes':likes, 'Reposts':reposts})
```

```python
df.describe()
```

```python
methods = ['spearman','kendall','pearson']
corr_value = {}
for method in methods:
    corr_value[method] = df.corr(method=method)["Likes"]['Reposts']
print(corr_value)
```

```python
df['RepostsCoef'] = df['Reposts'] * df['Likes'].mean()/df['Reposts'].mean()
```

```python
df.describe()
```

```python
df['Diff'] = round(df['Likes'][1:]-df['RepostsCoef'][1:])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df['Likes'], df['RepostsCoef'])
```
